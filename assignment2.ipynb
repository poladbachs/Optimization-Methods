**Exercise 1: Partial Derivatives, Gradient, and Hessian**

**Function Definition:**  
$$
f(x_1,x_2,x_3) = x_1\cos(x_3) + x_1x_3^2 + x_2\ln(x_2)
$$

---

**First-order Partial Derivatives:**

 $$\frac{\partial f}{\partial x_1} = \cos(x_3) + x_3^2$$
 $$\frac{\partial f}{\partial x_2} = \ln(x_2) + 1$$  
  *(using the derivative $ \frac{d}{dx}(x\ln(x)) = \ln(x) + 1$)*
 $$\frac{\partial f}{\partial x_3} = -x_1\sin(x_3) + 2x_1x_3$$

Thus, the gradient is:  
$$
\nabla f(x) = \begin{pmatrix}
\cos(x_3) + x_3^2\\[1mm]
\ln(x_2) + 1\\[1mm]
-x_1\sin(x_3) + 2x_1x_3
\end{pmatrix}.
$$

---

**Second-order Partial Derivatives (Hessian):**

 $$\frac{\partial^2 f}{\partial x_1^2} = 0$$
 $$\frac{\partial^2 f}{\partial x_1\partial x_2} = 0$$
 $$\frac{\partial^2 f}{\partial x_1\partial x_3} = -\sin(x_3) + 2x_3$$

 $$\frac{\partial^2 f}{\partial x_2^2} = \frac{1}{x_2}$$
 $$\frac{\partial^2 f}{\partial x_2\partial x_3} = 0$$

 $$\frac{\partial^2 f}{\partial x_3^2} = -x_1\cos(x_3) + 2x_1$$

*Note: Since the mixed partials are symmetric, the Hessian matrix is:*

$$
H(x) =
\begin{pmatrix}
0 & 0 & -\sin(x_3)+2x_3 \\[1mm]
0 & \frac{1}{x_2} & 0 \\[1mm]
-\sin(x_3)+2x_3 & 0 & -x_1\cos(x_3)+2x_1
\end{pmatrix}.
$$


**Exercise 2: Convex and Univariate Functions**

1. **First-order Condition:**  
   If $f$ is continuously differentiable, then $f$ is convex if and only if  
   $$
   f(y) \ge f(x) + f'(x)(y-x) \quad \forall\, x,y\in\mathbb{R}.
   $$
   **Graphical interpretation:**  
   Every tangent line to the function lies below the graph of $f$.

2. **Second-order Condition:**  
   If $f$ is twice continuously differentiable, then $f$ is convex if and only if  
   $$
   f''(x) \ge 0 \quad \forall\, x\in\mathbb{R}.
   $$

3. **Local Minimizer:**  
   For a convex function, any local minimizer is also a global minimizer.

4. **Stationary Point:**  
   If $f$ is convex and $x^\ast$ is a stationary point (i.e., $f'(x^\ast)=0$), then $x^\ast$ is a global minimizer.


**Exercise 3: Strictly Convex and Univariate Functions**

1. **Strict Convexity via Second Derivative:**  
   If $f : \mathbb{R} \to \mathbb{R}$ is twice continuously differentiable and  
   $$
   f''(x) > 0, \quad \forall x \in \mathbb{R},
   $$  
   then $f$ is strictly convex.

   **Counterexample (Converse is false):**  
   Consider $f(x) = x^4$. Although \(f\) is strictly convex, its second derivative is  
   $$
   f''(x)=12x^2, \quad \text{and } f''(0)=0.
   $$

2. **Uniqueness of Global Minimizer:**  
   If $f$ is strictly convex, then there exists at most one global minimizer.

**Exercise 4: Strongly Convex and Univariate Functions**

1. **Strong Convexity Condition:**  
   A function $f$ is called $\mu$-strongly convex (with $\mu > 0$) if for all $x,y \in \mathbb{R}$,
   $$
   f(y) \ge f(x) + f'(x)(y-x) + \frac{\mu}{2}(y-x)^2.
   $$
   **Graphical interpretation:**  
   The function lies above a quadratic function (parabola) with curvature $\mu$.

2. **Unique Global Minimizer:**  
   A $\mu$-strongly convex function has a unique global minimizer.


**Exercise 5: Smooth and Univariate Functions**

1. **L-smoothness:**  
   A function $f$ is L-smooth (i.e., its derivative is L-Lipschitz continuous) if for all $x,y\in\mathbb{R}$,
   $$
   f(y) \le f(x) + f'(x)(y-x) + \frac{L}{2}(y-x)^2.
   $$

2. **Relation between Strong Convexity and Smoothness:**  
   If $f$ is both $\mu$-strongly convex and L-smooth, then it must hold that  
   $$
   \mu \le L.
   $$

**Exercise 2: k-th Order Derivative of $ \sin(x) $**

The derivatives of $ \sin(x) $ cycle every 4 orders:

$$
f^{(k)}(x) = \sin\left(x + \frac{k\pi}{2}\right)
$$

Define a function that computes this derivative.

```python
def kth_derivative_sin(x, k):
    """
    Returns the k-th derivative of sin(x), using the cyclic property:
    f^(k)(x) = sin(x + k*pi/2)
    """
    return np.sin(x + (k * np.pi / 2))

# Example: compute 5th derivative at x = 1.0
print("5th derivative at x=1:", kth_derivative_sin(1.0, 5))
```

**Exercise 3: Taylor Approximation at $ \bar{x} = 0 $: Tangent and Higher Order**

1. Tangent Line (First-Order Taylor Approximation):

The first-order Taylor approximation (tangent line) at $ x = 0 $ is:

$$
t_1(x;0) = f(0) + f'(0) x = 0 + 1 \cdot x = x
$$

---

2. General k-th Order Taylor Approximation:

The k-th order Taylor series approximation at $ x = 0 $ is given by:

$$
t_k(x;0) = \sum_{i=0}^{k} \frac{f^{(i)}(0)}{i!} x^i
$$

For $ \sin(x) $, the derivatives at $ x = 0 $ follow a cyclic pattern:

$ f(0) = 0 $
$ f'(0) = 1 $
$ f''(0) = 0 $
$ f'''(0) = -1 $, etc.

Below, we compute and plot the approximations for $ k = 1, 3, 5, \dots, 19 $ along with the sine function.

```python
import math

def taylor_approx_sin(x, k):
    """
    Computes the k-th order Taylor approximation of sin(x) at x=0.
    """
    approx = 0
    for i in range(k+1):
        # f^(i)(0) = sin(i*pi/2)
        deriv = np.sin(i * np.pi / 2)
        approx += deriv * (x**i) / math.factorial(i)
    return approx

# Plot sine and its Taylor approximations
plt.figure(figsize=(10, 6))
plt.plot(x, f(x), 'k', linewidth=2, label='sin(x)')

# Plot approximations for k = 1, 3, 5, ... , 19
for k in range(1, 20, 2):
    approx_vals = taylor_approx_sin(x, k)
    plt.plot(x, approx_vals, label=f'Taylor order {k}')

plt.xlabel('x')
plt.ylabel('Function value')
plt.title('Taylor Approximations of sin(x) at x=0')
plt.legend()
plt.grid(True)
plt.show()
```

**Exercise 4: Approximation Error at $ x = 2 $**

To analyze the accuracy of the Taylor series approximation, we compute the absolute error:

$$
\left| \sin(2) - t_k(2;0) \right|
$$

for different orders $ k = 1, 3, 5, \dots, 19 $.

The results are displayed in a table, showing how the approximation improves as $ k $ increases.

```python
errors = []
ks = list(range(1, 20, 2))
true_value = f(2)

for k in ks:
    approx_value = taylor_approx_sin(2, k)
    error = np.abs(true_value - approx_value)
    errors.append(error)

# Print table of errors
print("k-th Order Taylor Approximation Error at x = 2:")
print("-------------------------------------------------")
print("Order (k)\tError")
for k, err in zip(ks, errors):
    print(f"{k:9d}\t{err:.4e}")
```

Comment: As $k$ increases, the Taylor approximation error decreases. The table shows a rapid decrease in error, indicating convergence of the Taylor series to $sin(2)$ as higher order terms are included.

#### **Problems 2: Gradient Descent for a Bivariate Function**

We now consider the function:

$$
f(x_1, x_2) = 100 (x_2 - x_1^2)^2 + (x_1 - 1)^2.
$$


**Exercise 1: Function Definition for f(x)**

We define a Python function to compute $ f(x) $ for any $ x \in \mathbb{R}^2 $.


```python
import numpy as np
import matplotlib.pyplot as plt

def f_rosenbrock(x):
    """
    Computes f(x) = 100*(x2 - x1^2)^2 + (x1 - 1)^2
    where x is a 2D numpy array [x1, x2].
    """
    x1, x2 = x[0], x[1]
    return 100 * (x2 - x1**2)**2 + (x1 - 1)**2
```

**Exercise 2: Contour Plot of $ f(x) $ and Zoomed-In View**

We now visualize the function $ f(x) $ over the domain $ [-2,2] \times [-2,2] $ using a contour plot.

A second plot provides a **zoomed-in view near the minimum** to help visually determine the optimal values.

```python
# Create a grid for contour plotting
x1_vals = np.linspace(-2, 2, 400)
x2_vals = np.linspace(-2, 2, 400)
X1, X2 = np.meshgrid(x1_vals, x2_vals)
Z = f_rosenbrock(np.array([X1, X2]))

# Contour plot for the full range
plt.figure(figsize=(8, 6))
contour = plt.contour(X1, X2, Z, levels=50, cmap='viridis')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Contour Plot of f(x) on [-2, 2] x [-2, 2]')
plt.colorbar(contour)
plt.show()

# Zoom in near the minimum (expected near (1,1))
plt.figure(figsize=(8, 6))
x1_zoom = np.linspace(0.5, 1.5, 400)
x2_zoom = np.linspace(0.5, 1.5, 400)
X1z, X2z = np.meshgrid(x1_zoom, x2_zoom)
Z_zoom = f_rosenbrock(np.array([X1z, X2z]))
contour_zoom = plt.contour(X1z, X2z, Z_zoom, levels=50, cmap='viridis')
plt.xlabel('x1')
plt.ylabel('x2')
plt.title('Zoomed Contour Plot near the Minimum')
plt.colorbar(contour_zoom)
plt.show()
```

**Exercise 4: Analytical Gradient and Its Function**

The function is:

$$
f(x_1, x_2) = 100 (x_2 - x_1^2)^2 + (x_1 - 1)^2.
$$

**Gradient Computation:**

Taking partial derivatives:

$$
\frac{\partial f}{\partial x_1} = -400 x_1 (x_2 - x_1^2) + 2 (x_1 - 1)
$$

$$
\frac{\partial f}{\partial x_2} = 200 (x_2 - x_1^2)
$$

Thus, the gradient is:

$$
\nabla f(x) =
\begin{pmatrix}
-400 x_1 (x_2 - x_1^2) + 2 (x_1 - 1) \\[1mm]
200 (x_2 - x_1^2)
\end{pmatrix}
$$

Now, we implement this as a function.


```python
def grad_f_rosenbrock(x):
    """
    Computes the gradient of f(x) for x in R^2.
    Returns a numpy array [df/dx1, df/dx2].
    """
    x1, x2 = x[0], x[1]
    df_dx1 = -400 * x1 * (x2 - x1**2) + 2 * (x1 - 1)
    df_dx2 = 200 * (x2 - x1**2)
    return np.array([df_dx1, df_dx2])
```

**Exercise 5: Gradient Descent Algorithm**

We implement **Gradient Descent** to iteratively minimize $ f(x) $.

**Inputs:**
- Function $ f(x) $ and its gradient $ \nabla f(x) $
- Step size $ \alpha $
- Initial point $ x_{\text{start}} $
- Max iterations (default: 1000)
- Stopping threshold for the gradient norm (default: $ 10^{-6} $)

The function returns all iterates of $ x $.


```python
import numpy as np

def gradient_descent(f, grad_f, x_start, alpha, max_iter=1000, tol=1e-6):
    """
    Performs gradient descent.
    
    Parameters:
      f       : function to minimize.
      grad_f  : gradient of f.
      x_start : initial point (numpy array).
      alpha   : step size.
      max_iter: maximum number of iterations.
      tol     : tolerance for the norm of the gradient.
      
    Returns:
      iterates: list of iterates.
    """
    x_current = np.array(x_start, dtype=float)
    iterates = [x_current.copy()]
    
    for _ in range(max_iter):
        grad = grad_f(x_current)
        if np.linalg.norm(grad) < tol:
            break
        x_current = x_current - alpha * grad
        iterates.append(x_current.copy())
    
    return iterates

# Given inputs
x0 = np.array([0.5, 2.0])  # Initial point
alpha = 0.001  # Step size (as given in the assignment)

# Run Gradient Descent with given inputs
iters = gradient_descent(f_rosenbrock, grad_f_rosenbrock, x0, alpha)

# Print final iterate
print("Last iterate:", iters[-1])
```
